```python
from PIL import ImageColor
from PIL import Image, ImageFilter
import binascii
import cv2 
from math import log
from math import e
from scipy import stats
from numpy import asarray
from math import floor
import numpy as np 
from google.colab.patches import cv2_imshow  
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
image = cv2.imread('star.jpg') 
image = cv2.resize(image, (430, 430))
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
 cv2_imshow(gray)
```

```python
row,col = gray.shape
height = int(row/10)
width = int(col/10)
entropy = np.zeros([height, width])
stdev = np.zeros([height, width])
```

```python
#entropy,stdev = entropyanalisis(height,width,gray,10)
```

```python

for i in range(height):
  for j in range(width):
    lst=[]
    for k in range(i*10,i*10+10):
      for l in range(j*10,j*10+10):
        lst.append(gray[k,l])
    stdev[i,j]= log((np.std(np.array(lst))),2)
    unique, frequency = np.unique(np.array(lst), return_counts = True)
    prob=frequency/100
    entropy[i,j]=stats.entropy(prob,base=2)

```

```python
image = cv2.imread('lena.png') 
#image = cv2.resize(image, (430, 430))
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
cv2_imshow(gray)
```

```python
row,col = gray.shape
height = int(row/5)
width = int(col/5)
entropy = np.zeros([height, width])
stdev = np.zeros([height, width])
```

```python
for i in range(height):
  for j in range(width):
    lst=[]
    for k in range(i*5,i*5+5):
      for l in range(j*5,j*5+5):
        lst.append(gray[k,l])
    stdev[i,j]= log((np.std(np.array(lst))),2)
    unique, frequency = np.unique(np.array(lst), return_counts = True)
    prob=frequency/25
    entropy[i,j]=stats.entropy(prob,base=2)
```

```python
c=plt.imshow(entropy, cmap='hot',vmin = np.min(entropy), vmax = np.max(entropy) , interpolation='nearest')
plt.colorbar(c)
plt.title('  entropy  ', fontweight ="bold")
plt.show()
```
